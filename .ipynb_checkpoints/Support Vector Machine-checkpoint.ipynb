```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.preprocessing import StandardScaler
```

```python
#https://www.kaggle.com/datasets/samuelcortinhas/credit-card-approval-clean-data
dataset = pd.read_csv("clean_dataset.csv")
```

```python
ethnicity = pd.get_dummies(dataset.Ethnicity)
ethnicity
```

```python
new_dataset = pd.concat([dataset.Approved,ethnicity],axis='columns')
new_dataset.groupby('Approved').mean()
```

```python
industry = pd.get_dummies(dataset.Industry)
new_dataset = pd.concat([dataset.Approved,industry],axis='columns')
new_dataset.groupby('Approved').mean()
```

```python
citizen = pd.get_dummies(dataset.Citizen)
new_dataset = pd.concat([dataset.Approved,citizen],axis='columns')
new_dataset.groupby("Approved").mean()
```

```python
features = dataset.drop(['DriversLicense','ZipCode', 'Married','Ethnicity','Industry','Citizen','Approved'], axis='columns')
features = StandardScaler().fit_transform(features)
labels = dataset.Approved
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=42)
```

```python
model = svm.SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_train_predictions = model.predict(X_train)
training_accuracy = accuracy_score(X_train_predictions,y_train)
training_accuracy
```
